```python
import seaborn as sns
sns.get_dataset_names()
```

```python
tip=sns.load_dataset("tips")
print(tip)
#print(tip.head(10))
```

```python
sns.scatterplot(x="tip",y="total_bill",data=tip,hue='day')
```

```python
sns.histplot(tip['tip'],kde=True,bins=10)
```

```python
sns.distplot(tip['tip'],kde=True,bins=10)
```

```python
sns.boxplot(x="day",y="tip",data=tip,hue="sex")
```

```python
sns.stripplot(x="day",y="tip",data=tip,hue="sex")
```

```python
# assign x and y to create a scatterplot (using scatterplot()) 
#with marginal histograms (using histplot()):
#kind` must be one of ['scatter', 'hist', 'hex', 'kde', 'reg', 'resid']
sns.jointplot(x="tip",y="total_bill",data=tip,kind="reg")
```

```python
sns.jointplot(x="tip",y="total_bill",data=tip)
```

```python
sns.jointplot(x="tip",y="total_bill",data=tip,kind="kde",shade=True)
```
